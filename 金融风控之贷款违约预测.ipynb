# 三、特征工程（数据处理）

## 3.1处理工作年限特征

```python
data_train['employmentLength']
```

```python
data_train["isDefault"].value_counts()

def employmentLength_to_int(s):
    if type(s) == float:
        return s
    if pd.isnull(s):
        return s
    else:
        return np.int8(s.split()[0])

data_train['employmentLength'].replace(to_replace='10+ years', value='10 years', inplace=True)
data_train['employmentLength'].replace('< 1 year', '0 years', inplace=True)
data_train['employmentLength'] =data_train['employmentLength'].apply(employmentLength_to_int)
data_train['employmentLength'].value_counts(dropna=False).sort_index()


data_test['employmentLength'].replace(to_replace='10+ years', value='10 years', inplace=True)
data_test['employmentLength'].replace('< 1 year', '0 years', inplace=True)
data_test['employmentLength'] =data_test['employmentLength'].apply(employmentLength_to_int)
data_test['employmentLength'].value_counts(dropna=False).sort_index()
```

```python
from sklearn.preprocessing import LabelEncoder

data_train["grade"] = LabelEncoder().fit_transform(data_train["grade"])
data_train["subGrade"] = LabelEncoder().fit_transform(data_train["subGrade"])

data_train["employmentLength"]
```

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer() #实例化，默认用均值填补
employmentLength = data_train.loc[:, "employmentLength"].values.reshape(-1, 1)

imp_mean = imp_mean.fit_transform(employmentLength)

data_train.loc[:, "employmentLength"] = imp_mean
data_train.loc[:, "employmentLength"]
```

```python
import datetime
data_train_new = data_train.copy()
data_train_new['issueDate'] = pd.to_datetime(data_train_new['issueDate'],format='%Y-%m-%d')
data_train_new["earliesCreditLine"] = pd.to_datetime(data_train_new["earliesCreditLine"],format='%Y-%m-%d')

startdate = datetime.datetime.strptime('2007-06-01', '%Y-%m-%d')
data_train_new['issueDate'] =data_train_new['issueDate'].apply(lambda x: x-startdate).dt.days
data_train_new['earliesCreditLine'] =data_train_new['earliesCreditLine'].apply(lambda x: x-startdate).dt.days
data_train_new.info()
```

```python
data_train_new.shape
```

```python
for i in data_train_new.columns[1:]:
    if len(data_train_new[i].isnull().value_counts()) == 2:
        data_train_new[i] = data_train_new[i].fillna(data_train[i].median())
        
for i in data_train_new.columns[1:]:
    if len(data_train_new[i].isnull().value_counts()) == 2:
        data_train_new[i] = data_train_new[i].fillna(data_train[i].median())
data_train_new.info()
```

```python
from sklearn.preprocessing import StandardScaler
transfer = StandardScaler()

data_train_new1 = data_train_new.copy()
data_train_new1.drop("isDefault", axis=1, inplace=True)

Y_train = data_train_new[["isDefault"]].values
X_train = data_train_new1.values

x_train = transfer.fit_transform(X_train)
# x_test = transfer.fit_transform(X_train)
```

# 四：建模及评价

## 第一种算法： 逻辑回归

```python
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x_train ,Y_train, test_size = 0.4 )
```

```python
from sklearn.linear_model import LogisticRegression as LR
lr = LR()
lr = lr.fit(x_train, y_train)
y_pred = lr.predict(x_val)
```

```python
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score

FPR, TPR, thresholds = roc_curve(y_val, y_pred)
plt.title("ROC")
plt.plot(FPR, TPR, "b")
plt.plot([0,1], [0, 1], 'r--')
plt.ylabel("TPR")
plt.xlabel("FPR")
```

```python
print("lr_AUC score:", roc_auc_score(y_val, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn = knn.fit(x_train, y_train)
y_pred = knn.predict(x_val)

FPR, TPR, thresholds = roc_curve(y_val, y_pred)
plt.title("ROC")
plt.plot(FPR, TPR, "b")
plt.plot([0,1], [0, 1], 'r--')
plt.ylabel("TPR")
plt.xlabel("FPR")
```

```python
from sklearn import metrics
import lightgbm as lgb
# x_train, x_val, y_train, y_val = train_test_split(x_train ,Y_train, test_size = 0.4 )
# X_train_split, X_val, y_train_split, y_val = train_test_split(x_train ,Y_train, test_size = 0.2 )

train_matrix = lgb.Dataset(x_train, label=y_train)
valid_matrix = lgb.Dataset(x_val, label=y_val)


params = {
            'boosting_type': 'gbdt',
            'objective': 'binary',
            'metric': 'auc',
            'learning_rate': 0.01,
            'num_leaves': 14,
            'max_depth': 19,
            'min_data_in_leaf': 37,
            'min_child_weight':1.6,
            'bagging_fraction': 0.98,
            'feature_fraction': 0.69,
            'bagging_freq': 96,
            'reg_lambda': 9,
            'reg_alpha': 7,
            'min_split_gain': 0.4,
            'nthread': 8,
            'seed': 2020,
            'silent': True,
}
# params = {
#             'boosting_type': 'gbdt',
#             'objective': 'binary',
#             'learning_rate': 0.1,
#             'metric': 'auc',
#             'min_child_weight': 1e-3,
#             'num_leaves': 31,
#             'max_depth': -1,
#             'reg_lambda': 0,
#             'reg_alpha': 0,
#             'feature_fraction': 1,
#             'bagging_fraction': 1,
#             'bagging_freq': 0,
#             'seed': 2020,
#             'nthread': 8,
#             'silent': True,
#             'verbose': -1,
# }

"""使用训练集数据进行模型训练"""
model = lgb.train(params, train_set=train_matrix, valid_sets=valid_matrix, num_boost_round=20000, verbose_eval=1000, early_stopping_rounds=200)

val_pre_lgb = model.predict(x_val, num_iteration=model.best_iteration)
fpr, tpr, threshold = metrics.roc_curve(y_val, val_pre_lgb)
roc_auc = metrics.auc(fpr, tpr)
print('未调参前lightgbm单模型在验证集上的AUC：{}'.format(roc_auc))
"""画出roc曲线图"""
plt.figure(figsize=(8, 8))
plt.title('Validation ROC')
plt.plot(fpr, tpr, 'b', label = 'Val AUC = %0.4f' % roc_auc)
plt.ylim(0,1)
plt.xlim(0,1)
plt.legend(loc='best')
plt.title('ROC')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
# 画出对角线
plt.plot([0,1],[0,1],'r--')
plt.show()
```
